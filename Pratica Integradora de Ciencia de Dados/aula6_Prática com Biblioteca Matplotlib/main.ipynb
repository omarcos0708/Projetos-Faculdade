```python
# 1. Importando as bibliotecas necessrias
import matplotlib.pyplot as plt
import numpy as np

# 2. Criando o conjunto de dados
x = np.linspace(0, 10, 100)  # 100 valores entre 0 e 10
y1 = np.sin(x)  # Funo seno
y2 = np.cos(x)  # Funo cosseno

# 3. Grfico de linha (seno)
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='sen(x)', color='blue', linewidth=2)
plt.title('Grfico da Funo Seno')
plt.xlabel('Valores de x')
plt.ylabel('sen(x)')
plt.grid(True)
plt.legend()
plt.show()

# 4. Grfico de disperso (cosseno)
plt.figure(figsize=(8, 5))
plt.scatter(x, y2, label='cos(x)', color='red', marker='o', alpha=0.6)
plt.title('Grfico de Disperso da Funo Cosseno')
plt.xlabel('Valores de x')
plt.ylabel('cos(x)')
plt.grid(True)
plt.legend()
plt.show()

# 5. Grfico de barras
valores = np.array([15, 30, 45, 10, 25])
rotulos = ['A', 'B', 'C', 'D', 'E']

plt.figure(figsize=(8, 5))
plt.bar(rotulos, valores, color=['skyblue', 'orange', 'lightgreen', 'pink', 'violet'])
plt.title('Grfico de Barras Personalizado')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 7. Salvando o grfico de barras
plt.savefig('grafico_barras.png', dpi=300, bbox_inches='tight')
plt.show()

# 6. Subplot combinado
plt.figure(figsize=(10, 8))

# Primeiro subplot (linha - seno)
plt.subplot(2, 1, 1)  # 2 linhas, 1 coluna, posio 1
plt.plot(x, y1, color='blue')
plt.title('Subplot 1: Funo Seno')
plt.grid(True)

# Segundo subplot (disperso - cosseno)
plt.subplot(2, 1, 2)  # 2 linhas, 1 coluna, posio 2
plt.scatter(x, y2, color='red', alpha=0.6)
plt.title('Subplot 2: Funo Cosseno')
plt.grid(True)

plt.tight_layout()  # Ajusta o espaamento entre subplots

# 8. Exibindo todos os grficos
plt.show()
```
