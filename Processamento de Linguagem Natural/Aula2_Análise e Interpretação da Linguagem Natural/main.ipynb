```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

### Texto para a analise

```python
texto = """
Com a evolução contínua da tecnologia, o processamento de linguagem natural (PLN) tornou-se uma ferramenta essencial na análise de grandes volumes de dados textuais. 
O PLN permite que máquinas compreendam e interpretem a linguagem humana, facilitando a extração de informações valiosas e a automação de tarefas que antes exigiam intervenção humana. 
Técnicas avançadas de PLN são utilizadas em diversas aplicações, desde assistentes virtuais até sistemas de recomendação, contribuindo significativamente para a eficiência e a personalização dos serviços oferecidos. 
À medida que o campo avança, novas metodologias e algoritmos são desenvolvidos para aprimorar a precisão e a eficácia do PLN, tornando-o uma área de estudo fascinante e em constante evolução.
"""
```

## 1. Pré-processamento, Tokenização e Normalização de Texto

```python
tokens = word_tokenize(texto.lower())

stop_words = set(stopwords.words('portuguese'))
tokens_sem_stopwords = [token for token in tokens if token.isalpha() and token not in stop_words]

lemmatizer = WordNetLemmatizer()
tokens_lematizados = [lemmatizer.lemmatize(token) for token in tokens_sem_stopwords]
```

```python
print("Tokens Lematizados:")
print(tokens_lematizados)
```

## 2. Análise de Palavras-Chave

```python
frequencia = Counter(tokens_lematizados)

print("\nFrequência das Palavras-Chave:")
for palavra, contagem in frequencia.most_common(10):
    print(f"{palavra}: {contagem}")
```

## 3. Representação Vetorial de Palavras

```python
modelo_w2v = Word2Vec(sentences=[tokens_lematizados], vector_size=100, window=5, min_count=1, workers=4)
```

```python
print("\nVetor da palavra 'processamento':")
print(modelo_w2v.wv['processamento'])
```

## 4. Visualização de Palavras-Chave

```python
palavras = list(frequencia.keys())
contagens = list(frequencia.values())

plt.figure(figsize=(10, 5))
plt.bar(palavras, contagens)
plt.xlabel('Palavras')
plt.ylabel('Frequência')
plt.title('Frequência das Palavras-Chave')
plt.xticks(rotation=45)
plt.show()
```
